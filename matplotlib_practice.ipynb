```python
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create the plot and add a label for the legend.
plt.figure(figsize=(20,14))
plt.plot(x_axis, y_axis, label='Boston', lw=2, marker='D')
# create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')
# set the y limit between 0 and 45.
plt.ylim(0, 45)
# create a title.
plt.title('PyBer Fare by Month')
# add the legend.
plt.legend()
plt.grid()
```

```python
# create the bar graph
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color='green', label='Boston')
# Create labels for the x and y axes.
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Create a title.
plt.title('PyBer Fare by Month')
# Add the legend.
plt.legend()
```

```python
# create a horizontal bar chart
plt.barh(y_axis, x_axis)
```

## Create a Vertical Bar Chart

```python
# create the horizontal bar plot invert y-axis
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k', label='New York')
```

```python
# create the horizontal bar plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# create the horizontal bar plot with ax.plt(), switch axes
fig, ax = plt.subplots(figsize=(20,14))
ax.barh(x_axis, y_axis, color='cyan')
ax.invert_yaxis()

ax.set_ylabel('Month')
ax.set_xlabel('Fare ($)')
ax.set_title('Fare by Month')
ax.legend('New York', loc='best')
ax.grid(True)
```
